```python
df = pd.read_csv('2022_audience.csv')
df
```

```python
df_home = df[df['연고지']==True]
df_home['구장'].value_counts()
```

```python
stadiums = ['PayPay 돔', '베루나 돔', '라쿠텐 생명 파크', '교세라 돔','ZOZO 마린',
            '삿포로 돔', '요코하마 스타디움','도쿄 돔','진구','마쓰다 스타디움', '반테린 돔', '고시엔']
```

```python
stadium_average = []
```

```python
for stadium in stadiums:
  stadium_data = df_home[df_home['구장'] == stadium]
  average = stadium_data['관중수'].mean()
  stadium_average.append(average)
stadium_average
```

```python
plt.figure(figsize=(20,8))
plt.bar(stadiums,stadium_average)
plt.title('구장별 관중 수')
plt.show()
```

```python
days= ['월', '화', '수', '목', '금', '토', '일']

day_average = []

for day in days:
  day_data = df_home[df_home['요일']==day]
  average = day_data['관중수'].mean()
  day_average.append(average)

day_average
plt.plot(days,day_average)
plt.title('요일별 관중 수')
plt.show()
```

```python
central_teams=['T','C','DB','G','S','D']
```

```python
central_numbers=[]

for home in central_teams:
  home_data = df[df['H팀']== home]
  lst =[]
  for visitor in central_teams:
    home_visitor_data = home_data[home_data['V팀']==visitor]
    average = home_visitor_data['관중수'].mean()
    lst.append(average)
  central_numbers.append(lst)
central_numbers
```

```python
plt.figure(figsize=(8,8))
plt.imshow(central_numbers, cmap='OrRd')

plt.colorbar()
plt.title('대전 카드별 관중 수(센트럴리그) ')
plt.xticks([0,1,2,3,4,5], central_teams)
plt.yticks([0,1,2,3,4,5], central_teams)
plt.xlabel('어웨이 팀')
plt.ylabel('홈 팀')
plt.show()
```
